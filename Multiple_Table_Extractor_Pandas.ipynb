```python
import pandas as pd
pd.set_option('display.max_rows',None)
```

```python
# Building Fuction 
def table_extracter(path):

    try:
        df = pd.read_csv(path,error_bad_lines=False, skip_blank_lines=False,header=None, warn_bad_lines=False)
        p = -2
        df2 = []
        for i in range(1,len(df)):
            #print(str(df.iloc[i,0]) == 'nan')
            if str(df.iloc[i,0]) == 'nan':
                df1 = df.loc[(p+1):(i-1),:]

                try:
                    df1.columns = df1.iloc[0]
                    df1 = df1[1:]
                    df1 = df1.dropna(axis=1, how='all')
                    df2.append(df1)
                    p = i
                except:
                    pass
                
        return df2
            
    except:
        print('Something went wrong')
```

```python
# Giving path to Function
df = table_extracter('data.csv')
```
